```python
df = pd.read_csv('/Users/liqi/Desktop/IE5374/作业1/COVID-19 Nursing Home Data 09.11.2022.csv')
df.head()
```

```python
# 1.Generate a table showing the distribution of the number of nursing home facilities by state and county
df_number = pd.DataFrame(df.groupby(['Provider State', 'County'])['Provider Name'].nunique().reset_index(name = 'Number'))
display(df_number)
```

```python
# 2.List the top 5 states by the number of nursing home facilities
df_top5 = pd.DataFrame(df.groupby(['Provider State'])['Provider Name'].nunique().reset_index())
df_top5.columns = ['State', 'Number']
df_top5 = df_topfive.sort_values(by = ['Number'], ascending = False).iloc[0:5]
display(df_top5)
```

```python
# 3.Conduct a facility analysis on shortage of aides, clinical staff, nursing staff and other staff. Define the following metric for this analysis: number of weeks the shortage was experienced. The metric should be categorized into 3 categories, 0-10, 11-20 and >20 weeks. 
df_st = pd.DataFrame(df.groupby(['Provider Name',
                                 'Shortage of Aides',
                                 'Shortage of Clinical Staff',
                                 'Shortage of Nursing Staff',
                                 'Shortage of Other Staff'])['Provider Name'].count().reset_index(name = 'Number'))
df_analysis = df_st.loc[(df_st['Shortage of Aides'] == 'Y') & 
                        (df_st['Shortage of Clinical Staff'] == 'Y') &
                        (df_st['Shortage of Nursing Staff'] == 'Y') & 
                        (df_st['Shortage of Other Staff'] == 'Y')]
df_analysis = df_analysis[['Provider Name', 'Number']]
weeks = []
for i in df_analysis['Number']:
    if i < 11:
        weeks.append('0-10')
    elif 11 <= i < 20:
        weeks.append('11-20')
    else:
        weeks.append('>20')
df_analysis['Shortage of weeks'] = weeks
df_shortage = df_analysis[['Provider Name', 'Shortage of weeks']]
display(df_shortage)
```

```python
# 4.For Middlesex county in Massachusetts compute the hospital bed occupancy in percentage for each facility. How many facilities had 100 % of hospital bed occupancy for 3 weeks or more?
df_MA = df.loc[(df['Provider State'] == 'MA') & (df['County'] == 'Middlesex')]
df_MA = df_MA[['Provider Name', 'Number of All Beds', 'Total Number of Occupied Beds']]
df_MA['ratio'] = df_MA['Total Number of Occupied Beds'] / df_MA['Number of All Beds']
df_bed = df_MA.loc[df_MA['ratio'] == 1.]
df_100 = df_bed['Provider Name'].value_counts().reset_index()
df_100.columns = ['Provider Name', 'weeks']
display(df_100)
(df_100['weeks'] >= 3).sum()
```
